```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
path = "x_array.txt" # путь, конечно, не такой, но палить папки на диске я не хочу
data = pd.read_csv(path, sep=' ', header=None)
data.head(4)
```

```python
for n in range(len(data.columns)):
    data[n] = data[n].str.replace(',', '.').astype(float)

data.head(4)
```

```python
iterations = np.linspace(0, len(data.loc[:, 0]), num=len(data.loc[:, 0]))
fig, ax = plt.subplots()
for i in range(len(data.columns)):
    ax.plot(iterations, data.loc[:, i])
ax.set_xlabel(r'$iter$')
ax.set_ylabel(r'$x$')
ax.set_title("График изменения независимых переменных")
plt.show()
```
